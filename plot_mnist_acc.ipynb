```python
x = np.linspace(-1,1,200)


err_128 = np.sqrt((1-x**2)/128)
err_512 = np.sqrt((1-x**2)/512)

plt.figure(figsize=(8, 6))
plt.plot(x, err_128, label='seq_len = 128', linestyle='-',)
plt.plot(x, err_512, label='seq_len = 512',  linestyle='--',)
plt.xlabel('The Value of x')
plt.ylabel('Mean Error of Approxing x with Random Sequences')
plt.grid(True)
plt.legend()
plt.show()
```
